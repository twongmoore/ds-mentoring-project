In section two we'll visualise our dataset. You need to consider what could effect who survived the titanic sinking based on the attributes we have. 

We use two extra packages here. Seaborn which contains our base plotting functions and matplotlib which has our customisations. An example script has been created below.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

Problem 1 - Create a bar chart counting the number of survivors vs deceased.

```python
# Create a count plot using Seaborn
sns.countplot(data=df, x='Survived')

# Set the labels and title
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival Count')

# Show the plot
plt.show()
```

Problem 2 - Create a bar chart counting the number of survivors vs deceased grouped by sex.

```python
# Set the style of the plot
sns.set_style("whitegrid")

# Create the countplot with hue
sns.countplot(data=df, x="Survived", hue="Sex")

# Set the labels and title
plt.xlabel("Survived")
plt.ylabel("Count")
plt.title("Survival Count by Sex")

# Display the plot
plt.show()
```

Problem 3 - Create a bar chart counting the number of survivors vs deceased grouped by class.

```python
# Set the style of the plot
sns.set_style("whitegrid")

# Create the countplot with hue
sns.countplot(data=df, x="Survived", hue="Pclass")

# Set the labels and title
plt.xlabel("Survived")
plt.ylabel("Count")
plt.title("Survival Count by Sex")

# Display the plot
plt.show()
```

Problem 3 - Visualise the mean and interquartile range of ages on the titanic by sex  

```python
# Set the style of the plot
sns.set_style("whitegrid")

# Create the violin plot
sns.violinplot(data=df, x="Sex", y="Age")

# Set the labels and title
plt.xlabel("Sex")
plt.ylabel("Age")
plt.title("Age Distribution by Sex")

# Display the plot
plt.show()
```

Problem 4 - Visualise whether fares and survival are related.

```python
# Set the style of the plot
sns.set_style("whitegrid")

# Create the box plot
sns.boxplot(data=df, x="Survived", y="Fare")

# Set the y-axis limit
plt.ylim(0, 300)

# Set the labels and title
plt.xlabel("Survived")
plt.ylabel("Fare")
plt.title("Fare Distribution by Survival")

# Display the plot
plt.show()
```

Problem 5 - Visualise whether Age and Survival are related.

```python
# Set the style of the plot
sns.set_style("whitegrid")

# Create the swarm plot
sns.swarmplot(data=df, x="Survived", y="Age", size=4)

# Set the labels and title
plt.xlabel("Survived")
plt.ylabel("Age")
plt.title("Age Distribution by Survival")

# Display the plot
plt.show()
```

Conclusion - Fare, gender and class appear to be variables that inform whether someone survives. Age is not.
